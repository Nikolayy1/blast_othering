```python
import json
import pandas as pd

bodies = []

with open("../full_comments/DePi_comments.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        obj = json.loads(line)
        body = obj.get("body")

        if body is not None:
            bodies.append(body)

# Create DataFrame (one column = one Excel column)
df = pd.DataFrame(bodies, columns=["body"])

# Save to Excel
df.to_excel("../full_comments/DePi_bodies.xlsx", index=False)
```

```python
import pandas as pd
import json

# Load Excel file, sheet 2 (0-based index â†’ sheet_name=1)
df = pd.read_excel("../full_comments/DePi_bodies.xlsx", sheet_name=1)

# Get text column, ensure strings, drop empty rows if needed
text_column = df["text"].dropna().astype(str)

# Build nested JSON structure
data = {
    str(i + 1): {"text": text}
    for i, text in enumerate(text_column)
}

# Save to JSON
with open("texts.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
